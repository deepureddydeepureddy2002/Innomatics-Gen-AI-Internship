```python
Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid
Real-Time Use
Telecom recharge apps
Payment gateways
Hint
Use function for validation
Use while loop for retry
```

```python
def validate_recharge():
    # Valid recharge plans
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
        except ValueError:
            print("Please enter a valid number.")
            continue

        # Rule 1: Minimum amount ₹50
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
            continue

        # Rule 2: Valid plans
        if amount not in valid_plans:
            print(f"Invalid plan. Available plans: {valid_plans}. Try again.")
            continue
        # If all checks pass
        print(f"Recharge of ₹{amount} is successful!")
        break  # Exit the loop
# Call the function
validate_recharge()
    
```

```python
Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise
Real-Time Use
Warehouse systems
Retail management
Hint
Use function
Use for loop on dictionary
```

```python
def check_inventory(inventory):
    for product, stock in inventory.items():
        if stock < 15:
            print(f"{product}: Reorder Alert! (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# Example inventory
inventory = {
    "Pens": 10,
    "Notebooks": 25,
    "Markers": 5,
    "Staplers": 20
}

# Call the function
check_inventory(inventory)
```

```python
 Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average ≥ 50 else Fail
Real-Time Use
Online exam systems
Evaluation portals
Hint
Use loop to calculate total
Return result from function
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark  # sum all marks
    
    average = total / len(marks)  # calculate average
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
student_marks = [60, 70, 45, 50]
result = process_result(student_marks)
print("Result:", result)
```

```python
Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50
₹12 per km
Peak hour → 25% extra
Allow retry using while loop
Real-Time Use
Ride-hailing apps
Transport systems
Hint
Function for fare calculation
While loop for retry
```

```python
def calculate_fare(distance, peak_hour=False):
    base_fare = 50
    per_km = 12
    fare = base_fare + (distance * per_km)
    
    if peak_hour:
        fare *= 1.25  # 25% extra
    
    return fare

# Retry loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        if distance <= 0:
            print("Distance must be positive. Try again.")
            continue
    except ValueError:
        print("Please enter a valid number.")
        continue

    peak = input("Is it peak hour? (yes/no): ").lower()
    peak_hour = True if peak == "yes" else False

    fare = calculate_fare(distance, peak_hour)
    print(f"Estimated Fare: ₹{fare:.2f}")
    break  # exit loop after successful calculation
```

```python
Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if ≥ 75% else Not Eligible
Real-Time Use
HR payroll systems
Internship tracking
Hint
Use loop to count present days
Use return statement
```

```python
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day.upper() == "P":  # count present days
            present_days += 1

    attendance_percentage = (present_days / total_days) * 100

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance)
print("Attendance Result:", result)
```

```python
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)
Real-Time Use
Login systems
Security validation
Hint
Use loop over string
Use conditions inside function
```

```python
def check_password(password):
    # Rule 1: Minimum length 8
    if len(password) < 8:
        return "Weak: Password too short"

    # Initialize flags
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    # Check rules
    if not has_digit:
        return "Weak: Must contain at least one digit"
    if not has_special:
        return "Weak: Must contain at least one special character (@#$)"

    return "Strong Password"


# Example usage
password = "MyPass@123"
result = check_password(password)
print(result)
```
